```python
import pandas as pd
from pydub import AudioSegment
import numpy as np
import matplotlib.pyplot as plt
```

```python
sound = AudioSegment.from_mp3('/data2/yamad/examples/54.mp3')

data = np.array(sound.get_array_of_samples())

x = data[::sound.channels]

plt.plot(x[::10])
plt.grid()
plt.show()
```

```python
print('samping rate: {}'.format(sound.frame_rate))
print('duration: {}'.format(sound.duration_seconds))
```
